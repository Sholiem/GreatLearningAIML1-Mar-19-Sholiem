```python
pd_data.drop('name',axis=1,inplace=True)
pd_data.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from scipy.stats import zscore

from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score, auc, roc_curve
```

```python
X = pd_data.drop("status", axis=1)
y = pd_data["status"]
```

```python
X = X.apply(zscore)  # convert all attributes to Z scale 
```

```python
from sklearn.tree import DecisionTreeClassifier

test_size = 0.30 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
dt_model = DecisionTreeClassifier(criterion = 'entropy')
dt_model.fit(X_train, y_train)
```

```python
y_predict = dt_model.predict(X_test)
```

```python
 metrics.confusion_matrix(y_test, y_predict)
```

```python
model_score = dt_model.score(X_test, y_test)
model_score
```

# Improve the model (Regularization) -------Iteration 2 ----------

```python
test_size = 0.30 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
dt_model = DecisionTreeClassifier(criterion = 'entropy', class_weight={0:.5,1:.5}, max_depth = 15, min_samples_leaf=5 )
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_test , y_test)
```

```python
y_predict = dt_model.predict(X_test)
```

```python
print(metrics.confusion_matrix(y_test, y_predict))
```
