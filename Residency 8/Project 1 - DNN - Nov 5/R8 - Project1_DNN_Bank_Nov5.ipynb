# Distinguish the feature and target set 

```python
import tensorflow as tf

import keras 
from keras.models import Sequential 
from keras.layers import Dense, Dropout 
from keras.optimizers import SGD, Adam, RMSprop
from keras.layers import BatchNormalization
from keras import utils
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
    
X = bank_data.drop("Exited", axis=1)
y = bank_data["Exited"]
```

```python
for i in X.columns:
    if X[i].dtype == 'object':
        print(i)
        X[i]=pd.Categorical(X[i]).codes
        print(X[i])
```

```python
X.head(5)
```

# Divide the data set into Train and test sets

```python
test_size = 0.20 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed )
```

# Normalize the train and test data 

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Initialize & build the model 

```python
model = Sequential() 
model.add(Dense(activation = 'relu', input_dim = 10, units=6, kernel_initializer='uniform'))
```

```python
model.add(Dense(activation = 'relu', units=6, kernel_initializer='uniform'))
model.add(Dense(activation = 'sigmoid', units=1, kernel_initializer='uniform')) 
model.summary() 
```

```python
# compile the model 
model.compile( loss ='binary_crossentropy', optimizer = Adam(), metrics =['accuracy']) 
```

```python
# train the model 
model.fit( X_train, y_train, batch_size = 100, epochs = 100) 
```

```python
# evaluate the model and print the accuracy score 

scores = model.evaluate( X_test, y_test) 

print('\ n loss:', scores[ 0]) 

print('\ n accuracy:', scores[ 1])
```

# The initial model gives an accuracy of 83.2%

# Optimize the model 

```python
model = Sequential() 
model.add(Dense(activation = 'relu', input_dim = 10, units=6, kernel_initializer='uniform'))
model.add(Dense(activation = 'relu', units=6, kernel_initializer='uniform'))
# Dropout
model.add(Dropout(0.1))
model.add(BatchNormalization()) 

model.add(Dense(activation = 'relu', units=6, kernel_initializer='uniform'))

model.add(Dropout(0.1))
model.add(BatchNormalization()) 

model.add(Dense(activation = 'sigmoid', units=1, kernel_initializer='uniform')) 
model.summary() 
```

```python
# compile the model 
model.compile( loss ='binary_crossentropy', optimizer = RMSprop(lr=0.001, rho=0.9)
, metrics =['accuracy'])
```

```python
# train the model 
model.fit( X_train, y_train, batch_size = 100, epochs = 100) 
```

```python
# evaluate the model and print the accuracy score 

scores = model.evaluate( X_test, y_test) 

print('\ n loss:', scores[ 0]) 

print('\ n accuracy:', scores[ 1])
```

# The Model was optimized to give an accuracy of 85.75% after tuning the architecture and optimizers parameters

# Predict the results using 0.5 as a threshold 

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
y_pred = (y_pred > 0.5)
print(y_pred)
```

# Print the Accuracy score and confusion matrix 

```python
confmat = confusion_matrix(y_test, y_pred)
print(confmat)
```
