# Q:: Is saving money (finances) gender dependant?

# Null Hypothesis = Finances are not gender dependent                                      #Alternate Hypothesis = Finances are gender dependent. 

```python
groupGender = relevantdf.groupby('Gender')
groupGender['Finances'].value_counts()
```

```python
from scipy import stats 
from scipy.stats import chi2_contingency

chi_arrayGender = np.array([[134,101,76,60,35],[218,148,95,64,63]])

chi2_statistic, p_value,dof , contingency   =  chi2_contingency(chi_arrayGender)
print(chi2_statistic, p_value, dof , contingency)
```

```python
print("chi2_statistic=",chi2_statistic)
print("p_value=",p_value)
```

# p_value = 0.196 > 0.05 . Hence we fail to reject the Null Hypothesis. Which concludes that Finances are not Gender Dependent

```python
chi_arrayVillagetown = np.array([[94,83,41,40,37],[258,166,57,131,87]])

chi2_statistic, p_value,dof , contingency   =  chi2_contingency(chi_arrayVillagetown)
print(chi2_statistic, p_value, dof , contingency)
```
